# Instagram-Photo-Collage

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Pictures=['bottom_left.jpg','bottom_right.jpg','top_left.jpg','top_right.jpg','center.jpeg']
normal=[]

# Showing images
def show(normal):
    for i in range(len(Pictures)):
        plt.subplot(int('23'+str(i+1)))
        plt.imshow(normal[i])
        plt.axis('off')

```

```python
# Converting images from bgr2rgb
for i in range(len(Pictures)):
    normal.append(cv2.cvtColor(cv2.imread(Pictures[i]),cv2.COLOR_BGR2RGB))
plt.figure(figsize=(20,10))    
show(normal)
normal[-1].shape
```

```python
# Reshaping image
for i in range(len(normal)-1):
    normal[i]=cv2.resize(normal[i],(200,200))
normal[-1]=cv2.resize(normal[-1],(100,100))    
    
plt.figure(figsize=(20,10))
show(normal)    
```

```python
frame=np.zeros([430,430,3], np.uint8)
plt.imshow(frame)
plt.show()
center_frame=np.zeros([120,120,3], np.uint8)
plt.imshow(center_frame)
```

```python
center_frame[10:110,10:110]=normal[-1]
plt.imshow(center_frame)
```

```python
frame[10:210,10:210]=normal[2]
frame[10:210,220:420]=normal[3]
frame[220:420,10:210]=normal[0]
frame[220:420,220:420]=normal[1]
frame[155:275,155:275]=center_frame
plt.figure(figsize=(25,10))
plt.imshow(frame)
plt.axis("off")
```

```python
frame = cv2.cvtColor(frame,cv2.COLOR_RGB2BGR)

cv2.imwrite('final_img.jpeg',frame)
```
